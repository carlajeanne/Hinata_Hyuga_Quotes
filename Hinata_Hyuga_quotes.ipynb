```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://shinobi-striker.fandom.com/wiki/Hinata_Hyuga/Quotes'

def extract_text_from_url(url):
    # Send a GET request to the page
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Create BeautifulSoup object; parse with 'html.parser'
        soup = bs(response.text, 'html.parser')
        hinata_quotes1 = [] # define hinata_quotes

        related_quotes = soup.find_all('i')

        for hinata_quote in related_quotes:
          quote = hinata_quote.text.strip()
          hinata_quotes1.append([url, quote]) # append to hinata_quotes
        return hinata_quotes1
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None

hinata_quotes1 = extract_text_from_url(url) # call function and assign return value to hinata_quotes

hinata_df_1 = pd.DataFrame(hinata_quotes1, columns=['URL','Quote'])
hinata_df_1.to_csv('hinata_hyuga_quotes_1.csv', index=False)
hinata_df_1
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

url = 'https://thatonerule.com/rule/hinata-hyuga-quotes/'

def extract_text_from_url(url):
    # Send a GET request to the page
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Create BeautifulSoup object; parse with 'html.parser'
        soup = bs(response.text, 'html.parser')
        hinata_quotes2 = []

        # Find the div with class 'post-content'
        post_content_div = soup.find('div', class_='post-content')

        # Check if the div exists
        if post_content_div:
            # Find all <p> tags within this div
            for element in post_content_div.find_all('p'):
                quote = element.text.strip()
                if quote:  # Ensure there is text to avoid empty quotes
                    # Add quotation marks around the quote
                    quoted_text = f'"{quote}"'
                    hinata_quotes2.append([url, quoted_text])
        return hinata_quotes2
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return []

hinata_quotes2 = extract_text_from_url(url)  # Call function and assign return value to hinata_quotes

hinata_df_2 = pd.DataFrame(hinata_quotes2, columns=['URL','Quote'])
hinata_df_2.to_csv('hinata_hyuga_quotes_2.csv', index=False)
hinata_df_2
```

```python
url = "https://www.cbr.com/naruto-hinata-best-quotes/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers) # use requests.get
response.raise_for_status()

soup = bs(response.content, 'lxml')

hinata_quotes3 = []

related_hinata_quotes = soup.find_all('span')

for hinata_quote in related_hinata_quotes:
    quote = hinata_quote.text.strip()
    if quote and '"' in quote:
        hinata_quotes3.append([url, quote])


hinata_df_3 = pd.DataFrame(hinata_quotes3, columns=['URL','Quote'])
hinata_df_3.to_csv('hinata_hyuga_quotes_3.csv', index=False)
hinata_df_3
```

```python
url = "https://screenrant.com/naruto-hinatas-best-most-memorable-quotes/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers) # use requests.get
response.raise_for_status()

soup = bs(response.content, 'lxml')

hinata_quotes4 = []

related_hinata_quotes = soup.find_all('h3')

for hinata_quote in related_hinata_quotes:
    quote = hinata_quote.text.strip()
    if quote and '"' in quote:
        hinata_quotes4.append([url, quote])


hinata_df_4 = pd.DataFrame(hinata_quotes4, columns=['URL','Quote'])
hinata_df_4.to_csv('hinata_hyuga_quotes_4.csv', index=False)
hinata_df_4
```

```python
url = "https://animemotivation.com/hinata-hyuga-quotes/"
response = requests.get(url)
response.raise_for_status()

soup = bs(response.content, 'lxml')

hinata_hyuga_quotes_5 = []

related_hinata_quotes = soup.find_all('blockquote')

for hinata_quote in related_hinata_quotes:
    par_quote = hinata_quote.find('p')
    if par_quote:
        quote = par_quote.text.strip()
        if quote:
            quote = quote.replace("– Hinata Hyuga", "").strip()
            hinata_hyuga_quotes_5.append([url, quote])

hinata_df_5 = pd.DataFrame(hinata_hyuga_quotes_5, columns=['URL', 'Quote'])
hinata_df_5.to_csv('hinata_hyuga_quotes_5.csv', index=False)
hinata_df_5
```

```python
# Load the two CSV files into separate DataFrames
df_info_01 = pd.read_csv('hinata_hyuga_quotes_1.csv')
df_info_02 = pd.read_csv('hinata_hyuga_quotes_2.csv')
df_info_03 = pd.read_csv('hinata_hyuga_quotes_3.csv')
df_info_04 = pd.read_csv('hinata_hyuga_quotes_4.csv')
df_info_05 = pd.read_csv('hinata_hyuga_quotes_5.csv')

# Concatenate the DataFrames
df_combined = pd.concat([df_info_01, df_info_02, df_info_03, df_info_04, df_info_05], ignore_index=True)

# Save the combined DataFrame to a new CSV file
df_combined.to_csv('hinata_combined_info.csv', index=False)

# Print the combined DataFrame
print(df_combined)
```

```python
import pandas as pd
import re

# Load the CSV file
df = pd.read_csv('hinata_combined_info.csv')

def clean_quote(quote):
    # Remove unwanted characters and fix encoding issues
    quote = re.sub(r'[^\x00-\x7F]+', ' ', quote)  # Remove non-ASCII characters
    quote = quote.replace('â€™', "'").replace('â€œ', '"').replace('â€', '"')
    return quote.strip()

# Apply the cleaning function to the 'Quote' column
df['Quote'] = df['Quote'].apply(clean_quote)

# Save the cleaned DataFrame to a new CSV file
df.to_csv('hinata_combined_info_cleaned.csv', index=False)
df.to_csv('hinata_combined_info_cleaned.txt', sep = '\t', index=False)

# Print the cleaned DataFrame
print(df)
```
